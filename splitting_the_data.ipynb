```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df  = pd.read_csv('news_dataset.csv',encoding='iso-8859-1')
df.head()
```

```python
display(df.topic.unique())
display(df.topic.value_counts())
```

```python
df_env = df.loc[df.topic=='environment ']
df = df.loc[df.topic!='environment ']
# display(df)
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['topic'], random_state=42)
```

```python
train_df = pd.concat([train_df,df_env])
```

```python
test_df
```

```python
train_df.to_csv('news_dataset_train.csv',index=False)
test_df.to_csv('news_dataset_test.csv',index=False)
```
